#### Variable Assignment

```python
Training_Folder = 'EPOT_DATA'
Validation_Aug_Folder = 'Validation_augmented_data'
Experiment = 'Group'
```

#### Parse the training and validation augmented data

```python
# Initialize lists for training and validation
training = []; training_labels = []
validation_aug = []; validation_aug_labels = [] # validation augmented data

#------------------------------------------------------------------------------

for subdir, dirs, files in os.walk(rootdir):
    
    if Training_Folder.lower() in subdir.lower():
        for file in files:
            if file.endswith('.mat'):
                lbl = [file.split('_')[1], file.split('_')[2][:3]]

                # load mat-file
                mat = loadmat(os.path.join(subdir, file))
                # get the numpy data from the mat file
                mat = mat['num_data']
                # append data to the list
                training.append(mat)
                training_labels.append(lbl)

    elif Validation_Aug_Folder.lower() in subdir.lower():
        for file in files:
            if file.endswith('.mat'):
                lbl = [file.split('_')[1], file.split('_')[2][:3]]

                # load mat-file
                mat = loadmat(os.path.join(subdir, file))
                # get the numpy data from the mat file
                mat = mat['num_data']
                # append data to the list
                validation_aug.append(mat)
                validation_aug_labels.append(lbl)
```

#### Parse the experimental data

```python
# Initialize lists for training and validation
ex_data = []; ex_labels = []

#------------------------------------------------------------------------------

for subdir, dirs, files in os.walk(rootdir):
    
    if Experiment.lower() in subdir.lower():
        for file in files:
            if file.endswith('.txt'):
                
                with open(os.path.join(subdir, file), 'r') as f:
                    x = pd.read_csv(os.path.join(subdir, file), delimiter='\t', decimal=',', skiprows=8, header=None)
                    x = np.asarray(x, dtype=float)
                    ex_data.append(x)

len(training)
```

## Augmentation

```python
first_quadrant=[]; second_quadrant=[];third_quadrant=[];fourth_quadrant=[];
for i in range(0,len(training)):
    piezo=training[i]
    first_quadrant.append(piezo)
for i in range(0,len(first_quadrant)):
    piezo=first_quadrant[i]
    piezo[:,2]=piezo[:,1]
    piezo[:,3]=piezo[:,2]
    piezo[:,4]=piezo[:,3]
    piezo[:,1]=first_quadrant[i][:,4]
    second_quadrant.append(piezo)
for i in range(0,len(second_quadrant)):
    piezo=second_quadrant[i]
    piezo[:,2]=piezo[:,1]
    piezo[:,3]=piezo[:,2]
    piezo[:,4]=piezo[:,3]
    piezo[:,1]=second_quadrant[i][:,4]
    third_quadrant.append(piezo)
for i in range(0,len(third_quadrant)):
    piezo=third_quadrant[i]
    piezo[:,2]=piezo[:,1]
    piezo[:,3]=piezo[:,2]
    piezo[:,4]=piezo[:,3]
    piezo[:,1]=third_quadrant[i][:,4]
    fourth_quadrant.append(piezo)
```
